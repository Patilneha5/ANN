# **Gradient Descent demo implementation**

```python
import numpy as np
import matplotlib.pyplot as plt

# Sample dataset
X = np.array([1, 2, 3, 4, 5])
Y = np.array([2, 4, 5, 4, 5])

# Initialize parameters
m = 0.0
b = 0.0

# Hyperparameters
learning_rate = 0.01
iterations = 1000
N = len(X)

# Gradient Descent loop with iteration tracking
for epoch in range(iterations):
    # Predict y-values with current m and b
    y_pred = m * X + b

    # Calculate errors and MSE
    errors = y_pred - Y
    mse = (1/N) * np.sum(errors ** 2)  # Mean Squared Error

    # Print progress every 100 iterations
    if epoch % 100 == 0:
        print(f"Iteration {epoch}: m = {m:.4f}, b = {b:.4f}, MSE = {mse:.4f}")

    # Compute gradients
    gradient_m = (1/N) * np.dot(errors, X)
    gradient_b = (1/N) * np.sum(errors)

    # Update parameters
    m -= learning_rate * gradient_m
    b -= learning_rate * gradient_b

# Final MSE after all iterations
y_pred_final = m * X + b
mse_final = (1/N) * np.sum((y_pred_final - Y) ** 2)
print(f"\nFinal Result (Iteration {iterations}): m = {m:.4f}, b = {b:.4f}, MSE = {mse_final:.4f}")

# Plot the best-fit line
plt.scatter(X, Y, color='blue', label='Data Points')
plt.plot(X, y_pred_final, color='red', label=f'Best-Fit Line: y = {m:.2f}x + {b:.2f}')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Gradient Descent: Best-Fit Line')
plt.legend()
plt.grid(True)
plt.show()
```

# **Student Score Prediction**

```python
import numpy as np
import matplotlib.pyplot as plt

hoursStudied = np.array([1, 2, 3, 4, 5])
score = np.array([40, 60, 75, 85, 90])

m = 0.0
b = 0.0

learning_rate = 0.01
iterations = 1000
N = len(hoursStudied)

#Gradient Descent
for epoch in range(iterations):
    y_pred = m * hoursStudied + b

    errors = y_pred - score
    mse = (1/N) * np.sum(errors ** 2)

    if epoch % 100 == 0:
        print(f"Iteration {epoch}: m = {m:.4f}, b = {b:.4f}, MSE = {mse:.4f}")

    gradient_m = (1/N) * np.dot(errors, hoursStudied)
    gradient_b = (1/N) * np.sum(errors)

    m -= learning_rate * gradient_m
    b -= learning_rate * gradient_b

y_pred_final = m * hoursStudied + b
mse_final = (1/N) * np.sum((y_pred_final - score) ** 2)
print(f"\nFinal Result (Iteration {iterations}): m = {m:.4f}, b = {b:.4f}, MSE = {mse_final:.4f}")

#plot the Graph
plt.scatter(hoursStudied, score, color='blue', label='Data Points')
plt.plot(hoursStudied, y_pred_final, color='red', label=f'Best-Fit Line: y = {m:.2f}x + {b:.2f}')
plt.xlabel('Hours Studied')
plt.ylabel('Score')
plt.title('Student Score Prediction: Best-Fit Line')
plt.legend()
plt.grid(True)
plt.show()
```

# **Temperature Prediction**

```python
import numpy as np
import matplotlib.pyplot as plt

altitude = np.array([10, 15, 20, 25, 30])
temperature = np.array([25, 28, 30, 32, 40])

m = 0.0
b = 0.0

learning_rate = 0.001
iterations = 100000
N = len(altitude)

for epoch in range(iterations):
    y_pred = m * altitude + b

    errors = y_pred - temperature
    mse = (1/N) * np.sum(errors ** 2)
    #mse = np.mean(errors ** 2)  # Mean Squared Error

    if epoch % 10000 == 0:
        print(f"Iteration {epoch}: m = {m:.2f}, b = {b:.2f}, MSE = {mse:.2f}")

    gradient_m = (1/N) * np.sum(errors * altitude)
    gradient_b = (1/N) * np.sum(errors)

    m -= learning_rate * gradient_m
    b -= learning_rate * gradient_b


y_pred_final = m * altitude + b
# Corrected MSE calculation:
mse_final = (1/N) * np.sum((y_pred_final - temperature)**2)
#mse_final = np.mean((y_pred_final - temperature) ** 2)  # Mean Squared Error
print(f"\nFinal Result (Iteration {iterations}): m = {m:.2f}, b = {b:.4f}, MSE = {mse_final:.2f}")

#plot the Graph
plt.scatter(altitude, temperature, color='blue', label='Data Points')
plt.plot(altitude, y_pred_final, color='red', label=f'Best-Fit Line: y = {m:.2f}x + {b:.2f}')
plt.xlabel('Altitude')
plt.ylabel('temperature')
plt.title('Temperature Prediction : Best-Fit Line')
plt.legend()
plt.grid(True)
plt.show()
```

# **House Price Prediction**

```python
import numpy as np
import matplotlib.pyplot as plt

squareFeet = np.array([1000, 1500, 2000, 2500, 3000])
price = np.array([20000, 25000, 30000, 35000, 40000])

#

#Initialize Parameters
m = 0.0
b = 0.0

#Hyper Parameters
learning_rate = 0.001
iterations = 10000
N = len(squareFeet)

#Gradient Descent
for epoch in range(iterations):
    y_pred = m * squareFeet + b

    errors = y_pred - price
    mse = (1/N) * np.sum(errors ** 2)

    if epoch % 1000 == 0:
        print(f"Iterations {epoch} : m = {m:.2f}, b = {b:.2f}, MSE = {mse:.2f} ")

    gradient_m = (1/N) * np.sum(errors * squareFeet)
    gradient_b = (1/N) *np.sum(errors)

    m -= learning_rate * gradient_m
    b -= learning_rate * gradient_b



#Final Predicted Output and MSE
y_pred_final = m * squareFeet + b
mse_final = (1/N) * np.sum((y_pred_final - price) ** 2)


#plot the Graph
plt.scatter(squareFeet, price, color='blue', label='Data Points')
plt.plot(squareFeet, y_pred_final, color='red', label=f'Best-Fit Line: y = {m:.2f}x + {b:.2f}')
plt.xlabel('Square Feet')
plt.ylabel('Price')
plt.title('House Price Prediction : Best-Fit Line')
plt.legend()
plt.grid(True)
plt.show()
```
